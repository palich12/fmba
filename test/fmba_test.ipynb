# Подготовка данных

```python
import pandas as pd
import numpy as np
```

### загружаем данные

```python
data_raw = pd.read_excel("test_DS.xlsx")
```

```python
data_raw.info()
```

```python
data_raw.head()
```

### Перемешиваем данные

```python
data = data_raw.sample(frac=1).reset_index(drop=True)
```

```python
data.head()
```

### Разделяем на входной и выходной вектора

```python
_id = data["id"]
_id.head()
```

```python
_class = data["class"]
_class.head()
```

```python
_age = data["age"]
_age.head()
```

```python
X = data.drop(columns=["id", "class", "age"])
X.head()
```

### Разделяем на тестовую и учебную выборки

```python
test_size = int(X.shape[0]*0.1)
test_index = list(X.index[0:test_size])
len(test_index)
```

```python
train_id    = _id.drop(test_index)
train_class = _class.drop(test_index)
train_age   = _age.drop(test_index)
train_X     = X.drop(test_index)
train_X.head()
```

```python
test_id    = _id.loc[test_index]
test_class = _class.loc[test_index]
test_age   = _age.loc[test_index]
test_X     = X.loc[test_index]
test_X.head()
```

```python
test_class
```

# Нормируем входной вектор

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
```

```python
X_scaler = StandardScaler()
X_scaler.fit(train_X)
train_X_scaled = pd.DataFrame(X_scaler.transform(train_X))
test_X_scaled  = pd.DataFrame(X_scaler.transform(test_X))
```

```python
train_X_scaled.head()
```

```python
test_X_scaled.head()
```

# Определение класса с помощью градиентного бустинга

```python
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(train_X_scaled, train_class)
```

```python
predicted_class = gbc.predict(test_X_scaled)
```

```python
list(zip(test_class, predicted_class))
```

```python
success = 0
for p, t in zip(predicted_class, test_class):
    if p == t:
        success += 1
success / len(test_class)
```

Спрогнозировано с точностью 88%

# Определение возраста с помощью градиентного бустинга

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor()
gbr.fit(train_X_scaled, train_age)
```

```python
predicted_age = gbr.predict(test_X_scaled)
```

```python
list(zip(test_age, predicted_age))
```

```python
error = 0
for p, t in zip(predicted_age, test_age):
    error += abs(p-t) / max(test_age)
success = 1 - error / len(test_age)
success
```

Спрогнозировано с точностью 91%
